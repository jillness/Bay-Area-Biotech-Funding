```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("crunchbase_SF_privateCompanies_biotech.csv")
df.head()
```

```python
df.count()
```

```python
df = df.drop(['Organization/Person Name URL', 'Funding Status', 'Last Funding Amount ($)', 'Last Funding Type'], axis=1)
df.head()
```

```python
#df = df.dropna(how="any")
#df.head()
```

```python
df.count()
```

```python
df['Total Funding Amount ($)'] = pd.to_numeric(df['Total Funding Amount ($)'], errors='coerce').fillna(0)
df['Total Equity Funding Amount ($)'] = pd.to_numeric(df['Total Equity Funding Amount ($)'], errors='coerce').fillna(0)
df.head()
```

```python
df.dtypes
```

```python
df['Average Amount of Funding Per Round ($)'] = df['Total Funding Amount ($)']/df['Number of Funding Rounds']
df['Average Amount of Funding Per Investor ($)'] = df['Total Funding Amount ($)']/df['Number of Investors']
df['Equity Funding as a Percentage of Total Funding Amount'] = df['Total Equity Funding Amount ($)']/df['Total Funding Amount ($)']*100
df.head(100)
```

Here we loop to separate keywords into respective columns

```python
medical = []
pharma = []
diagnostics = []
genetics = []
neuro = []
healthcare = []
tech = []
other = []

for category in df['Categories']:
    
    if 'Medical' in category:
        medical.append(1)
    else:
        medical.append(0)
    if 'Pharmaceuticals and Therapeutics' in category:
        pharma.append(1)
    else:
        pharma.append(0)
    if 'Health Diagnostics' in category:
        diagnostics.append(1)
    else:
        diagnostics.append(0)
    if 'Genetics' in category:
        genetics.append(1)
    else:
        genetics.append(0)
    if 'Neuroscience' in category:
        neuro.append(1)
    else:
        neuro.append(0)
    if 'Health Care' in category:
        healthcare.append(1)
    else:
        healthcare.append(0)
    if 'Information Technology and Software' in category:
        tech.append(1)
    else:
        tech.append(0)
        
    if any(c in category for c in('Medical', 'Pharmaceuticals and Therapeutics', 'Health Diagnostics', 'Genetics', 'Neuroscience', 'Health Care', 'Information Technology and Software')):
        other.append(0)
    else:
        other.append(1)

        
```

```python
len(other)
```

```python
df['Medical'] = medical
df['Pharmaceuticals and Therapeutics'] = pharma
df['Health Diagnostics'] = diagnostics
df['Genetics'] = genetics
df['Neuroscience'] = neuro
df['Health Care'] = healthcare
df['Information Technology and Software'] = tech
df['Other'] = other
```
