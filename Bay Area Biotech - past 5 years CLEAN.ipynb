```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_2014 = pd.read_csv("Bay Area Biotech 2014 Data.csv")
df_2015 = pd.read_csv("Bay Area Biotech 2015 Data.csv")
df_2016 = pd.read_csv("Bay Area Biotech 2016 Data.csv")
df_2017 = pd.read_csv("Bay Area Biotech 2017 Data.csv")
df_2018 = pd.read_csv("Bay Area Biotech 2018 Data.csv")

df_list_raw = [df_2014, df_2015, df_2016, df_2017, df_2018]
```

```python
df_list = []

for df in df_list_raw:
    df = df.drop(['Organization/Person Name URL', 'Last Funding Type', 
                  'Last Equity Funding Type', 'Last Equity Funding Amount ($)', 
                  'Last Equity Funding Amount ($)', 'Total Equity Funding Amount ($)', 
                  'Last Funding Type', 'Last Funding Amount ($)'], axis=1)
    df['Total Funding Amount ($)'] = pd.to_numeric(df['Total Funding Amount ($)'], errors='coerce').fillna(0)
    df = df.dropna(how='any')
    df_list.append(df)
```

```python
[df_2014, df_2015, df_2016, df_2017, df_2018] = df_list
```

```python
#KEYWORDS:
#medical <- 
#pharma <- biopharma, pharmaceuticals, therapeutics
#diagnostics <- healthdiagnostics, 
#healthcare <- clinicaltrials
#genetics
#neuroscience
#other

df_list_clean = []

for df in df_list:
    
    medical = []
    pharma = []
    diagnostics = []
    genetics = []
    neuro = []
    healthcare = []
    tech = []
    other = []

    medical_funds = []
    pharma_funds = []
    diagnostics_funds = []
    genetics_funds = []
    neuro_funds = []
    healthcare_funds = []
    tech_funds = []
    other_funds = []
    
    reduced_cat = []
    
    i = 0
    
    year = df.iloc[1, 4]
    
    #need to take differences from year(n+1).fundingamount - year(n).fundingamount
    #replace totals if company was present in previous years
    
    
    for category in df['Categories']:
        
        cat_string = ''
        
        df['Date'] = f"1/1/{year}"

        if 'Medical' in category:
            medical.append(1)
            medical_funds.append(df.iloc[i, 3])
            cat_string += 'Medical, '
        else:
            medical.append(0)
            medical_funds.append(0)
            
        if any(c in category for c in('Pharmaceuticals', 'Therapeutics', 'Biopharma')):
            pharma.append(1)
            pharma_funds.append(df.iloc[i, 3])
            cat_string += 'Pharmaceuticals and Therapeutics, '
        else:
            pharma.append(0)
            pharma_funds.append(0)
            
        if 'Health Diagnostics' in category:
            diagnostics.append(1)
            diagnostics_funds.append(df.iloc[i, 3])
            cat_string += 'Health Diagnostics, '
        else:
            diagnostics.append(0)
            diagnostics_funds.append(0)
            
        if 'Genetics' in category:
            genetics.append(1)
            genetics_funds.append(df.iloc[i, 3])
            cat_string += 'Genetics, '
        else:
            genetics.append(0)
            genetics_funds.append(0)
            
        if 'Neuroscience' in category:
            neuro.append(1)
            neuro_funds.append(df.iloc[i, 3])
            cat_string += 'Neuroscience, '
        else:
            neuro.append(0)
            neuro_funds.append(0)
            
        if any(c in category for c in('Health Care', 'Clinical Trials')):
            healthcare.append(1)
            healthcare_funds.append(df.iloc[i, 3])
            cat_string += 'Health Care, '
        else:
            healthcare.append(0)
            healthcare_funds.append(0)

        if any(c in category for c in('Information Technology', 'Software', 'Analytics',
                                      'Big Data', 'Bioinformatics', 'Machine Learning', 'Cloud Computing')):
            tech.append(1)
            tech_funds.append(df.iloc[i, 3])
            cat_string += 'IT and Software, '
        else:
            tech.append(0)
            tech_funds.append(0)

        if any(c in category for c in('Medical', 
                                      'Pharmaceutical', 'Therapeutics', 'Biopharma', 
                                      'Health Diagnostics', 
                                      'Genetics', 
                                      'Neuroscience', 
                                      'Health Care', 'Clinical Trials',
                                      'Information Technology', 'Software', 'Big Data', 'Bioinformatics', 'Machine Learning', 'Cloud Computing')):
            other.append(0)
            other_funds.append(0)
        else:
            other.append(1)
            other_funds.append(df.iloc[i, 3])
            cat_string += 'Other, '

            
        i += 1
        reduced_cat.append(cat_string)
        

    df['Medical'] = medical
    df['Pharmaceuticals and Therapeutics'] = pharma
    df['Health Diagnostics'] = diagnostics
    df['Genetics'] = genetics
    df['Neuroscience'] = neuro
    df['Health Care'] = healthcare
    df['Information Technology and Software'] = tech
    df['Other'] = other

    df['Medical Funding ($)'] = medical_funds
    df['Pharmaceuticals and Therapeutics Funding ($)'] = pharma_funds
    df['Health Diagnostics Funding ($)'] = diagnostics_funds
    df['Genetics Funding ($)'] = genetics_funds
    df['Neuroscience Funding ($)'] = neuro_funds
    df['Health Care Funding ($)'] = healthcare_funds
    df['Information Technology and Software Funding ($)'] = tech_funds
    df['Other Funding ($)'] = other_funds
    
    df['Category Reduced'] = reduced_cat
    
    df_list_clean.append(df)

    
```

```python
[df_2014, df_2015, df_2016, df_2017, df_2018] = df_list_clean
```

```python
full_df = pd.concat(df_list_clean)
full_df = full_df.drop(['Unnamed: 10', 'Year'], axis=1)
```
